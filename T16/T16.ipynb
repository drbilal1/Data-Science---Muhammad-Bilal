```python
import matplotlib.pyplot as plt
```

Sample data

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
```

Basic Plots:

1. Create a simple line plot

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y1)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

2. Plot multiple lines on the same graph

```python
plt.figure(figsize=(5, 3))
plt.plot(x, x, label='X-Values')
plt.plot(x, y1, label='Prime Numbers')
plt.plot(x, y2, label='Even Numbers')
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

3. Add markers to a line plot

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y1, marker='o', label='Prime Numbers')
plt.title("Line Plot with Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

4. Customize line styles and colors

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y1, color='green', linestyle='--', linewidth=2, marker='s', label='Prime Numbers')
plt.plot(x, y2, color='blue', linestyle='-.', linewidth=2, marker='^', label='Even Numbers')
plt.title("Customized Line Styles and Colors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

Sample data

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]
```

Scatter Plots:

1. Create a basic scatter plot

```python
plt.figure(figsize=(5, 3))
plt.scatter(x1, y1)
plt.title("Basic Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

2. Customize marker styles, sizes, and colors

```python
plt.figure(figsize=(5, 3))
plt.scatter(x1, y1, color='red', marker='*', s=100, label='Stars')
plt.title("Custom Marker Styles, Sizes, and Colors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

3. Plot multiple scatter plots on the same graph

```python
plt.figure(figsize=(5, 3))
plt.scatter(x1, y1, color='green', marker='o', s=80, label='Prime Points')
plt.scatter(x2, y2, color='blue', marker='^', s=80, label='Even Points')
plt.title("Multiple Scatter Plots")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

4. Add labels and title to the scatter plot

```python
plt.figure(figsize=(5, 3))
plt.scatter(x1, y1, color='purple', marker='D', s=90)
for i in range(len(x1)):
    plt.text(x1[i] + 0.1, y1[i], f"({x1[i]},{y1[i]})", fontsize=9)
plt.title("Scatter Plot with Point Labels")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

Sample data

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 98]
```

Bar Charts:

1. Create a vertical bar chart

```python
plt.figure(figsize=(5, 3))
plt.bar(categories, values)
plt.title("Vertical Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True, axis='y')
plt.show()
```

2. Create a horizontal bar chart

```python
plt.figure(figsize=(5, 3))
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.grid(True, axis='x')
plt.show()
```

3. Customize bar colors and add edge colors

```python
plt.figure(figsize=(5, 3))
plt.bar(categories, values, color='skyblue', edgecolor='black', linewidth=1.5)
plt.title("Custom Bar Colors with Edges")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True, axis='y')
plt.show()
```

Sample data

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0.1, 0, 0)  # Explode the 1st slice (Apples)
```

Pie Charts:

1. Create a basic pie chart

```python
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
plt.show()
```

2. Customize slice colors and add a legend

```python
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, colors=colors)
plt.title('Pie Chart with Custom Colors and Legend')
plt.legend(labels, title="Fruits")
plt.show()
```

3. Explode a slice of the pie chart

```python
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True)
plt.title('Exploded Slice Pie Chart')
plt.show()
```

4. Add percentage labels to slices

```python
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%1.1f%%')
plt.title('Pie Chart with Percentage Labels')
plt.legend(labels, title="Fruits")
plt.show()
```
