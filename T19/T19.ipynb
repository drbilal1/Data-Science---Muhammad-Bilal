```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Linear Regression Plot – sns.regplot()

```python
df = sns.load_dataset("tips")
df
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
plt.title("Linear Regression: Total Bill vs Tip")
plt.grid(True)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title("Linear Regression by Gender")
plt.grid(True)
plt.show()
```

```python
sns.residplot(x='total_bill', y='tip', data=df)
plt.title("Residual Plot")
plt.grid(True)
plt.show()
```

```python
sns.histplot(df['total_bill'], bins=20, kde=False , color= 'skyblue')
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.grid(True)
plt.show()
```

```python
sns.kdeplot(df['total_bill'], shade=True, color='purple')
plt.title("KDE Plot of Total Bill")
plt.grid(False)
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")  # To ignore deprecation warning

sns.distplot(df['total_bill'], bins=20, color='teal')
plt.title("Distplot of Total Bill")
plt.grid(True)
plt.show()
```

```python
sns.distplot(df['tip'], bins=20, color='teal')
plt.title("Distplot of Tip")
plt.grid(True)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Boxplot: Total Bill by Day")
plt.grid(True)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violin Plot: Total Bill by Day")
plt.grid(True)
plt.show()
```

Pair Plot

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue='species')

plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
iris.head()
```

```python
#corr = iris.corr()
corr = iris.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap of Iris Correlation Matrix")
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex') # Other kinds: 'kde', 'reg', 'scatter'
plt.show()
```

```python
g = sns.FacetGrid(df, col='sex', row='time')
g.map(sns.histplot, 'total_bill')
plt.suptitle("Facet Grid: Total Bill Distribution by Sex and Time", y=1.03)
plt.show()
```
