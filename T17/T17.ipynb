```python
import matplotlib.pyplot as plt
import numpy as np
```

Sample data

```python
x = np.linspace(0, 10, 100)
y2 = np.cos(x)
y1 = np.sin(x)

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]

hist_data = np.random.randn(1000)
```

Subplots:

1. Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).


```python
# Create a 2x2 subplot figure
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Scatter plot
axs[0, 1].scatter(scatter_x, scatter_y, color='green')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X-values')
axs[0, 1].set_ylabel('Y-values')

# Bar plot
axs[1, 0].bar(categories, values, color='orange', edgecolor='black')
axs[1, 0].set_title('Bar Chart')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Value')

# Histogram
axs[1, 1].hist(hist_data, bins=20, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the figure
plt.show()
```

2. Customize the subplots with titles, axis labels, and legends.

```python
# Create a 2x2 subplot figure
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Line Plot
axs[0, 0].plot(x, y1, label='sin(x)', color='blue')
axs[0, 0].plot(x, y2, label='cos(x)', color='green', linestyle='--')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()

# Subplot 2: Scatter Plot
axs[0, 1].scatter(x, y1, label='sin(x)', color='purple', marker='o')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[0, 1].legend()

# Subplot 3: Bar Chart
axs[1, 0].bar(categories, values, color='orange', edgecolor='black', label='Values')
axs[1, 0].set_title('Bar Chart')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Value')
axs[1, 0].legend()

# Subplot 4: Histogram
axs[1, 1].hist(hist_data, bins=20, color='skyblue', edgecolor='black', label='Data')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

Histograms

1. Create a histogram for a given numerical data set. Customize the number of bins.

```python
data = np.random.randint(10, 100, size=50)

# Customize the number of bins
num_bins = 9

# Create histogram
plt.figure(figsize=(5, 3))
plt.hist(data, bins=num_bins, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')

# Show plot
plt.grid(True)
plt.show()
```

2. Plot multiple histograms on the same figure to compare different data sets.

```python
data1 = np.random.normal(loc=40, scale=5, size=500)
data2 = np.random.normal(loc=50, scale=10, size=500)
data3 = np.random.normal(loc=60, scale=5, size=500)

bins = 20

# Plot multiple histograms
plt.figure(figsize=(7, 4))
plt.hist(data1, bins=bins, alpha=0.5, label='Dataset 1', color='blue')
plt.hist(data2, bins=bins, alpha=0.5, label='Dataset 2', color='green')
plt.hist(data3, bins=bins, alpha=0.5, label='Dataset 3', color='red')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Multiple Histograms')
plt.legend()
plt.grid(True)

plt.show()
```

```python
data1 = np.random.normal(loc=40, scale=5, size=500)
data2 = np.random.normal(loc=50, scale=10, size=500)
data3 = np.random.normal(loc=60, scale=7, size=500)

bins = 20

plt.figure(figsize=(8, 5))
plt.hist(data1, bins=bins, density=True, alpha=0.5, label='Dataset 1', color='blue')
plt.hist(data2, bins=bins, density=True, alpha=0.5, label='Dataset 2', color='green')
plt.hist(data3, bins=bins, density=True, alpha=0.5, label='Dataset 3', color='red')

plt.xlabel('Value')
plt.ylabel('Density (Normalized Frequency)')
plt.title('Comparison of Normalized Histograms')
plt.legend()
plt.grid(False)

plt.show()
```

Box Plots


1.
Create a box plot for a given numerical data set.

```python
data = np.random.normal(loc=50, scale=10, size=100)

plt.figure(figsize=(6, 5))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='skyblue'))

#plt.grid(True)
plt.show()
```

2. Generate multiple box plots in a single figure to compare different data sets.


```python
data1 = np.random.normal(loc=40, scale=5, size=100)
data2 = np.random.normal(loc=50, scale=10, size=100)
data3 = np.random.normal(loc=60, scale=7, size=100)

# Combine data into a list
data = [data1, data2, data3]

plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='red'))
plt.show()
```

3.
Customize the box plots with titles, labels, and change the colors.

```python
data1 = np.random.normal(loc=40, scale=5, size=100)
data2 = np.random.normal(loc=50, scale=10, size=100)
data3 = np.random.normal(loc=60, scale=7, size=100)

# Combine data into a list
data = [data1, data2, data3]

plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor='blue'),
            medianprops=dict(color='yellow'))

# Add labels and title
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])
plt.ylabel('Value')
plt.title('Comparison of Multiple Box Plots')
plt.grid(True)

plt.show()
```
