```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('insurance.csv')
```

```python
df
```

```python
sns.scatterplot(x='sex', y='region', hue = 'smoker', data=df)
plt.title('Scatter Plot of Sex vs Region')
plt.grid(True)
plt.show()
```

```python
sns.set_style("dark")
sns.lineplot(x='age', y= 'charges', data=df, marker="o")
plt.title("Line Plot of Age vs Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.grid(True)
plt.show()
```

```python
sns.barplot(x='age', y='smoker', data=df)
plt.title("Avg Fare by Class")
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='region', data=df)
plt.title('Region')
plt.grid(True)
plt.show()
```
